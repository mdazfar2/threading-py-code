```python
import threading as th
import time

def program1():
    while True:
        print("I'm Azfar..")
        print()
        time.sleep(1)  

def program2():
    while True:
        print("I'm Suman..")
        print()
        time.sleep(1)  

if __name__ == "__main__":
    thread1 = th.Thread(target=program1)
    thread2 = th.Thread(target=program2)

    thread1.start()
    thread2.start()

    thread1.join()
    thread2.join()
```
